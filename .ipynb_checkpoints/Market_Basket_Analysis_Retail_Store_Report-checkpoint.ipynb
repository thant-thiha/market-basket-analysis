# Market Basket Analysis of a Local Retail Store

### Import Libraries and Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
import time
import warnings
warnings.filterwarnings('ignore')
```

### Load and Explore

```python
# Load datasets
transaction_data =  pd.read_csv('datasets/transaction_data.csv')
product = pd.read_csv('datasets/product.csv')

# Filter the data to keep only rows for the first 90 days (3 months) for computational efficiency
transaction_data = transaction_data[transaction_data['DAY'] <= 90]
```

```python
print("Transaction Data Shape:", transaction_data.shape)
print("\nTransaction Data Sample:")
print(transaction_data.head())
```

```python
print("Product Data Shape:", product.shape)
print("\nProduct Data Sample:")
print(product.head())
```

### Data Preprocessing

```python
# Merge transaction with product info
data = transaction_data.merge(product, on='PRODUCT_ID', how='left')

print(f"Merged data shape: {data.shape}")
print("\nMerged data sample:")
print(data.head(5).T)
```

```python
# Check for missing values
print("Missing values:")
print(data.isnull().sum())
```

```python
print("\nData types:")
print(data.dtypes)
```

### Exploratory Data Analysis

```python
print(f"Unique Households: {data['household_key'].nunique()}")
print(f"Unique Products: {data['PRODUCT_ID'].nunique()}")
print(f"Total Transactions: {len(data)}")
print(f"Average basket size: {transaction_data.groupby('BASKET_ID').size().mean():.2f} items")
print(f"Median basket size: {transaction_data.groupby('BASKET_ID').size().median():.0f} items")
print(f"Date Range: Week {data['WEEK_NO'].min()} to Week {data['WEEK_NO'].max()}")
```

```python
# Top products by sales value
print("Top 10 Products by Sales Value:")
top_products = data.groupby('PRODUCT_ID')['SALES_VALUE'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# Top departments
print("Top 10 Departments by Sales Value:")
top_departments = data.groupby('DEPARTMENT')['SALES_VALUE'].sum().sort_values(ascending=False).head(10)
print(top_departments)
```

```python
# Visualize purchase patterns
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Sales by Department
dept_sales = data.groupby('DEPARTMENT')['SALES_VALUE'].sum().sort_values(ascending=False).head(15)
axes[0, 0].barh(range(len(dept_sales)), dept_sales.values)
axes[0, 0].set_yticks(range(len(dept_sales)))
axes[0, 0].set_yticklabels(dept_sales.index)
axes[0, 0].set_xlabel('Total Sales Value')
axes[0, 0].set_title('Top 15 Departments by Sales')
axes[0, 0].invert_yaxis()

# Purchases over time
weekly_sales = data.groupby('WEEK_NO')['SALES_VALUE'].sum()
axes[0, 1].plot(weekly_sales.index, weekly_sales.values)
axes[0, 1].set_xlabel('Week Number')
axes[0, 1].set_ylabel('Total Sales Value')
axes[0, 1].set_title('Sales Trend Over Time')
axes[0, 1].grid(True)

# Transaction distribution by day
day_counts = data['DAY'].value_counts()
axes[1, 0].bar(day_counts.index, day_counts.values)
axes[1, 0].set_xlabel('Day')
axes[1, 0].set_ylabel('Number of Transactions')
axes[1, 0].set_title('Transactions by Day')

# Quantity distribution
com_sales = data.groupby('COMMODITY_DESC')['SALES_VALUE'].sum().sort_values(ascending=False).head(15)
axes[1, 1].barh(range(len(com_sales)), com_sales.values)
axes[1, 1].set_yticks(range(len(com_sales)))
axes[1, 1].set_yticklabels(com_sales.index)
axes[1, 1].set_xlabel('Total Sales Value')
axes[1, 1].set_title('Top 15 Commodities by Sales')
axes[1, 1].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# Visualize basket size distribution
basket_sizes = transaction_data.groupby('BASKET_ID').size()

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(basket_sizes, bins=50, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Basket Size (Number of Items)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of Basket Sizes')
axes[0].axvline(basket_sizes.mean(), color='red', linestyle='--', label=f'Mean: {basket_sizes.mean():.2f}')
axes[0].legend()

# Top 20 most frequent products
top_products = transaction_data['PRODUCT_ID'].value_counts().head(20)
axes[1].barh(range(len(top_products)), top_products.values)
axes[1].set_yticks(range(len(top_products)))
axes[1].set_yticklabels(top_products.index)
axes[1].set_xlabel('Frequency')
axes[1].set_title('Top 20 Most Purchased Products')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

### Prepare Data for Market Basket Analysis

```python
# Create basket-product matrix
# Group by basket and get list of products
basket_product = data.groupby(['BASKET_ID', 'COMMODITY_DESC'])['QUANTITY'].sum().unstack().fillna(0)

# Convert to binary (0/1) format - if product was purchased or not
basket_product_binary = basket_product.applymap(lambda x: 1 if x > 0 else 0)

print("Basket-Product Matrix Shape:", basket_product_binary.shape)
print("\nSample of Basket-Product Binary Matrix:")
print(basket_product_binary.head())
```

```python
# Calculate sparsity
total_cells = basket_product_binary.shape[0] * basket_product_binary.shape[1]
non_zero_cells = (basket_product_binary > 0).sum().sum()
sparsity = 1 - (non_zero_cells / total_cells)
print(f"Matrix Sparsity: {sparsity:.2%}")
```

### Apply Apriori Algorithm

```python
# Set minimum support threshold
min_support = 0.01  # 1% - items appearing in at least 1% of transactions
min_confidence = 0.5  # 50% confidence
min_lift = 2.0  # Lift greater than 2.0

# Measure execution time for Apriori
start_time = time.time()
frequent_itemsets_apriori = apriori(basket_product_binary, 
                                     min_support=min_support, 
                                     use_colnames=True,
                                     verbose=0,low_memory=True)
apriori_time = time.time() - start_time

print(f"Apriori Execution Time: {apriori_time:.4f} seconds")
print(f"Number of Frequent Itemsets Found: {len(frequent_itemsets_apriori)}")
```

```python
# Display top frequent itemsets
frequent_itemsets_apriori_sorted = frequent_itemsets_apriori.sort_values('support', ascending=False)
print("\nTop 10 Frequent Itemsets (by Support):")
print(frequent_itemsets_apriori_sorted.head(10))
```

```python
# Generate association rules from Apriori
rules_apriori = association_rules(frequent_itemsets_apriori, 
                                   metric="confidence", 
                                   min_threshold=min_confidence)
```

```python
# Filter by lift
rules_apriori = rules_apriori[rules_apriori['lift'] >= min_lift]

print(f"\nNumber of Association Rules Generated: {len(rules_apriori)}")
print("\nTop 10 Association Rules (by Lift):")
rules_apriori_sorted = rules_apriori.sort_values('lift', ascending=False)
print(rules_apriori_sorted[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))
```

### Apply FP-Growth Algorithm

```python
# Measure execution time for FP-Growth
start_time = time.time()
frequent_itemsets_fpgrowth = fpgrowth(basket_product_binary, 
                                       min_support=min_support, 
                                       use_colnames=True,
                                       verbose=0)
fpgrowth_time = time.time() - start_time

print(f"FP-Growth Execution Time: {fpgrowth_time:.4f} seconds")
print(f"Number of Frequent Itemsets Found: {len(frequent_itemsets_fpgrowth)}")
```

```python
# Display top frequent itemsets
frequent_itemsets_fpgrowth_sorted = frequent_itemsets_fpgrowth.sort_values('support', ascending=False)
print("\nTop 10 Frequent Itemsets (by Support):")
print(frequent_itemsets_fpgrowth_sorted.head(10))
```

```python
# Generate association rules from FP-Growth
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, 
                                    metric="confidence", 
                                    min_threshold=min_confidence)

# Filter by lift
rules_fpgrowth = rules_fpgrowth[rules_fpgrowth['lift'] >= min_lift]

print(f"\nNumber of Association Rules Generated: {len(rules_fpgrowth)}")
print("\nTop 10 Association Rules (by Lift):")
rules_fpgrowth_sorted = rules_fpgrowth.sort_values('lift', ascending=False)
print(rules_fpgrowth_sorted[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))
```

### Performance Comparison

```python
comparison_df = pd.DataFrame({
    'Metric': ['Execution Time (seconds)', 
               'Frequent Itemsets Found', 
               'Association Rules Generated',
               'Speed Advantage'],
    'Apriori': [f"{apriori_time:.4f}", 
                len(frequent_itemsets_apriori), 
                len(rules_apriori),
                "Baseline"],
    'FP-Growth': [f"{fpgrowth_time:.4f}", 
                  len(frequent_itemsets_fpgrowth), 
                  len(rules_fpgrowth),
                  f"{(apriori_time/fpgrowth_time):.2f}x faster" if fpgrowth_time < apriori_time else f"{(fpgrowth_time/apriori_time):.2f}x slower"]
})

print("\n", comparison_df.to_string(index=False))
```

```python
# Visualize performance comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Execution time comparison
algorithms = ['Apriori', 'FP-Growth']
times = [apriori_time, fpgrowth_time]
colors = ['#FF6B6B', '#4ECDC4']

axes[0].bar(algorithms, times, color=colors, edgecolor='black', alpha=0.7)
axes[0].set_ylabel('Time (seconds)')
axes[0].set_title('Execution Time Comparison')
axes[0].set_ylim(0, max(times) * 1.2)
for i, (alg, t) in enumerate(zip(algorithms, times)):
    axes[0].text(i, t + max(times)*0.02, f'{t:.4f}s', ha='center', fontweight='bold')

# Frequent itemsets comparison
itemsets_count = [len(frequent_itemsets_apriori), len(frequent_itemsets_fpgrowth)]
axes[1].bar(algorithms, itemsets_count, color=colors, edgecolor='black', alpha=0.7)
axes[1].set_ylabel('Count')
axes[1].set_title('Frequent Itemsets Found')
axes[1].set_ylim(0, max(itemsets_count) * 1.2)
for i, (alg, count) in enumerate(zip(algorithms, itemsets_count)):
    axes[1].text(i, count + max(itemsets_count)*0.02, f'{count}', ha='center', fontweight='bold')

# Association rules comparison
rules_count = [len(rules_apriori), len(rules_fpgrowth)]
axes[2].bar(algorithms, rules_count, color=colors, edgecolor='black', alpha=0.7)
axes[2].set_ylabel('Count')
axes[2].set_title('Association Rules Generated')
axes[2].set_ylim(0, max(rules_count) * 1.2)
for i, (alg, count) in enumerate(zip(algorithms, rules_count)):
    axes[2].text(i, count + max(rules_count)*0.02, f'{count}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

### Rules Quality Comparison


```python
# Compare rule statistics
print("APRIORI RULES STATISTICS:")
print(rules_apriori[['support', 'confidence', 'lift']].describe())

print("\nFP-GROWTH RULES STATISTICS:")
print(rules_fpgrowth[['support', 'confidence', 'lift']].describe())
```

```python
# Visualize rules distribution
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

metrics = ['support', 'confidence', 'lift']
titles = ['Support', 'Confidence', 'Lift']

for i, (metric, title) in enumerate(zip(metrics, titles)):
    # Apriori
    axes[0, i].hist(rules_apriori[metric], bins=30, edgecolor='black', alpha=0.7, color='#FF6B6B')
    axes[0, i].set_xlabel(title)
    axes[0, i].set_ylabel('Frequency')
    axes[0, i].set_title(f'Apriori - {title} Distribution')
    axes[0, i].axvline(rules_apriori[metric].mean(), color='red', linestyle='--', 
                       label=f'Mean: {rules_apriori[metric].mean():.3f}')
    axes[0, i].legend()
    
    # FP-Growth
    axes[1, i].hist(rules_fpgrowth[metric], bins=30, edgecolor='black', alpha=0.7, color='#4ECDC4')
    axes[1, i].set_xlabel(title)
    axes[1, i].set_ylabel('Frequency')
    axes[1, i].set_title(f'FP-Growth - {title} Distribution')
    axes[1, i].axvline(rules_fpgrowth[metric].mean(), color='blue', linestyle='--', 
                       label=f'Mean: {rules_fpgrowth[metric].mean():.3f}')
    axes[1, i].legend()

plt.tight_layout()
plt.show()
```

### Verify Itemsets Consistency


```python
# Check if both algorithms found the same frequent itemsets
apriori_itemsets_set = set(frequent_itemsets_apriori['itemsets'].apply(frozenset))
fpgrowth_itemsets_set = set(frequent_itemsets_fpgrowth['itemsets'].apply(frozenset))

common_itemsets = apriori_itemsets_set.intersection(fpgrowth_itemsets_set)
apriori_only = apriori_itemsets_set - fpgrowth_itemsets_set
fpgrowth_only = fpgrowth_itemsets_set - apriori_itemsets_set

print(f"\nCommon Itemsets: {len(common_itemsets)}")
print(f"Apriori Only: {len(apriori_only)}")
print(f"FP-Growth Only: {len(fpgrowth_only)}")

if len(apriori_only) == 0 and len(fpgrowth_only) == 0:
    print("\nIDENTICAL")
else:
    print("\nDIFFERENT)")
```

### Top Association Rules Analysis

```python
top_n = 15

print(f"TOP {top_n} ASSOCIATION RULES ANALYSIS")

# Convert frozensets to strings for display
def format_rule(rule):
    antecedents = ', '.join(list(rule['antecedents']))
    consequents = ', '.join(list(rule['consequents']))
    return f"{antecedents}  {consequents}"

print("\nAPRIORI - Top Rules by Lift:")
top_apriori = rules_apriori_sorted.head(top_n).copy()
top_apriori['rule'] = top_apriori.apply(format_rule, axis=1)
print(top_apriori[['rule', 'support', 'confidence', 'lift']].to_string(index=False))

print("\n" + "="*80)
print("\nFP-GROWTH - Top Rules by Lift:")
top_fpgrowth = rules_fpgrowth_sorted.head(top_n).copy()
top_fpgrowth['rule'] = top_fpgrowth.apply(format_rule, axis=1)
print(top_fpgrowth[['rule', 'support', 'confidence', 'lift']].to_string(index=False))
```

```python
# Visualize top rules
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Apriori top rules
top_ap = rules_apriori_sorted.head(10).copy()
top_ap['rule_str'] = top_ap.apply(format_rule, axis=1)
axes[0].barh(range(len(top_ap)), top_ap['lift'].values, color='#FF6B6B', edgecolor='black', alpha=0.7)
axes[0].set_yticks(range(len(top_ap)))
axes[0].set_yticklabels(top_ap['rule_str'].values, fontsize=9)
axes[0].set_xlabel('Lift')
axes[0].set_title('Apriori - Top 10 Rules by Lift')
axes[0].invert_yaxis()

# FP-Growth top rules
top_fp = rules_fpgrowth_sorted.head(10).copy()
top_fp['rule_str'] = top_fp.apply(format_rule, axis=1)
axes[1].barh(range(len(top_fp)), top_fp['lift'].values, color='#4ECDC4', edgecolor='black', alpha=0.7)
axes[1].set_yticks(range(len(top_fp)))
axes[1].set_yticklabels(top_fp['rule_str'].values, fontsize=9)
axes[1].set_xlabel('Lift')
axes[1].set_title('FP-Growth - Top 10 Rules by Lift')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```
